```python
#line 
from matplotlib import pyplot as plt

x = [5, 2, 9, 7]

y = [10, 6, 8, 4]

plt.plot(x,y)
plt.show()
```

```python
#bar
plt.bar(x,y)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
labels = ['Std 1', 'Std 2', 'Std 3']
sizes = [10, 15, 20]

plt.pie(sizes, labels=labels)
plt.show()
```

```python
plt.pie(sizes, labels=labels)
plt.title("student data")
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.show()
```

```python
myexplode = [ 0.3, 0, 0]
plt.pie(sizes, labels=labels, explode = myexplode, shadow = True)
plt.show()
```

```python
myexplode = [ 0.3, 0, 0]
plt.pie(sizes, labels=labels, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```
